## Carregando os dados em Xarray

````python
# Analisa os arquivos em disco e agrupa em observações. Indique o número do grupo que deseja carregar.
# obs = uira.select_obs(group=0) - Com grupo pré-selecionado
obs = uira.select_obs() # solicita input do usuário
````

````python
# Agrupa os arquivos selecionados em chunks de tamanho threshold (MB) e carrega os dados dos chunks com índices indicados em chunks_idx
Xda = uira.load_fits(obs, chunks_idx=[0, 20], threshold=200)
````

````python
Xda.chunk(chunks={"Time":10000, "Freq":4096})
````

````python
# Resample em intervalo de 1 min.
DA_resample = Xda.resample(Time="1min").median()
````

````python
df = DA_resample.values
````
