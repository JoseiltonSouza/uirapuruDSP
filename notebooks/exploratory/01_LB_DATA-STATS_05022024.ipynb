# Características estatísticas do sinal

```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from astropy.io import fits
```

```python
files = list(Path("../../data/raw/").glob("*.fit"))
avgs = []
stds = []
for file in files[::10]:
    with fits.open(file) as hdul:
        data = hdul[0].data
        avg = np.mean(data, axis=0)
        std = np.std(data, axis=0)
        avgs.append(avg)
        stds.append(avg)
```

```python
df = pd.DataFrame({"Mean": np.mean(np.vstack(avgs), axis=0), "Std": np.std(np.vstack(avgs), axis=0)})
```

```python
df.to_csv("./ruido.csv", index=False)
```
